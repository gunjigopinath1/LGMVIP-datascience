## Importing packages

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

## Loading Image

```python
image = cv2.imread("tom.jpg")
cv2.imshow("Tom", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Converting the Original Image to GrayScale Image:

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap="gray")
plt.title(" Grayscale image:")
```

## Invert the Grayscale image

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image)
plt.title("Black And White Inverted")
```

## Blur the inverted image

```python
blur_img=cv2.GaussianBlur(inverted_image, (111,111),0)
plt.imshow(blur_img)
plt.title("blurred inverted grayscale image:")
```

## Convert to pencil sketch

```python
pencil_sketch = cv2.divide(gray_image, blur_img, scale=256.0)
plt.imshow(pencil_sketch,cmap="gray")
plt.title("Pencil sketch Image")
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(image)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Pencil Sketch', size=18)
rgb_sketch=cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
